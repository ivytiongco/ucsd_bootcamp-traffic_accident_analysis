```python
# Ivy

vehicles = cali2['Results'][0][0]['CrashResultSet']['Vehicles']
license_type = []
age = []
age_name = []
county = []
death_daname = []
doa_name = []
injury = []

for v in range(len(vehicles)):
    license_type.append(vehicles[v]['L_TYPENAME']),
    age.append(vehicles[v]['Persons'][0]['AGE']),
    age_name.append(vehicles[v]['Persons'][0]['AGENAME']),                    
    county.append(vehicles[v]['Persons'][0]['COUNTYNAME']),                    
    death_daname.append(vehicles[v]['Persons'][0]['DEATH_DANAME']),
    doa_name.append(vehicles[v]['Persons'][0]['DOANAME']),                    
    injury.append(vehicles[v]['Persons'][0]['INJ_SEVNAME'])                  
                        
    print(license_type)
    print(age)
    print(age_name)
    print(county)
    print(death_daname)
    print(doa_name)
    print(injury)
```

```python
# Ivy

data_df = pd.DataFrame({
    "License Type": license_type,
    "Age": age,
    "Age Name": age_name,
    "County": county,
    "Death DANAME": death_daname,
    "DOA Name": doa_name,
    "Injury": injury
})

data_df
```
